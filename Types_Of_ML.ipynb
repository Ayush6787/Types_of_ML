```python
import matplotlib.pyplot as plt
import numpy as np

# Data
np.random.seed(42)
class1 = np.random.multivariate_normal([2, 2], [[0.5, 0], [0, 0.5]], 50)
class2 = np.random.multivariate_normal([5, 5], [[0.5, 0], [0, 0.5]], 50)
boundary_x = np.linspace(0, 7, 100)
boundary_y = boundary_x

# Plot
plt.figure(figsize=(6, 6))
plt.scatter(class1[:, 0], class1[:, 1], color='blue', label='Class 1')
plt.scatter(class2[:, 0], class2[:, 1], color='red', label='Class 2')
plt.plot(boundary_x, boundary_y, color='black', linestyle='--', label='Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Classification Example')
plt.legend()
plt.grid()
plt.show()
```

```python
# Data
x = np.linspace(0, 10, 100)
y = 2 * x + np.random.normal(0, 1, 100)  # Linear trend with noise

# Plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, 2 * x, color='red', label='Regression Line')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Regression Example')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.datasets import make_blobs

# Generate clusters
centers = [[2, 2], [5, 5], [8, 2]]
cluster_data, cluster_labels = make_blobs(n_samples=150, centers=centers, cluster_std=1, random_state=42)

# Plot
plt.figure(figsize=(6, 6))
for i, color in enumerate(['red', 'blue', 'green']):
    plt.scatter(cluster_data[cluster_labels == i, 0], cluster_data[cluster_labels == i, 1], label=f'Cluster {i+1}', color=color)

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clustering Example')
plt.legend()
plt.grid()
plt.show()
```
